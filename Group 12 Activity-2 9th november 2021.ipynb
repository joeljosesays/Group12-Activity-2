```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('diabetes_model 11thnov.csv')
```

```python
df.isna().sum()
```

# 1) Create a Count plot of Outcome

```python
sns.countplot(x='Outcome',data=df)
```

# 2. Analyze the effect of age on Outcome using Boxplot

```python
sns.boxplot(x='Outcome',y='Age',data=df)
```

# 3. Check for the chances of getting diabetes if it is common in family, using the Daiabetes Pedigree Function

```python
df
```

```python
plt.figure(figsize=(8,8))
sns.boxplot(x='Outcome',y='DiabetesPedigreeFunction',data=df)
```

# 4. Check for the relationship with Insulin & Glucose and group the datapoints by Outcome.


```python
df
```

```python
plt.scatter(df['Insulin'],df['Glucose'],c=df['Outcome'])
plt.xlabel('Insulin')
plt.ylabel('Glucose')
```

```python
sns.heatmap(df.corr(),annot=True)
plt.xticks(rotation=90)
```
